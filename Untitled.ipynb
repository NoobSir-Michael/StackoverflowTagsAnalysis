```python
from bs4 import BeautifulSoup
import pandas as pd

def phase(htmlPath):
    fp = open(htmlPath).read ()
    soup = BeautifulSoup(fp)
    li_list = soup.find_all(class_ = "grid--item s-card js-tag-cell d-flex fd-column")
    valueDict = {"tagName":[], "info":[], "totalQuestion":[], "questionToday":[], "questionPrevious":[]}
    for li in li_list:
        # d-flex jc-space-between ai-center mb12
        tagName = li.find(class_="d-flex jc-space-between ai-center mb12").getText()
        tagName = tagName.strip()
        # flex--item fc-medium mb12 v-truncate4
        info = li.find(class_="flex--item fc-medium mb12 v-truncate4")
        if(info!=None):
            info = info.getText()
        else:
            info = ""
        info = info.strip()
        # mt-auto d-flex jc-space-between fs-caption fc-black-400 -> flex--item
        totalQuestion = li.find(class_="mt-auto d-flex jc-space-between fs-caption fc-black-400").find(class_="flex--item").getText()
        totalQuestion = totalQuestion.strip()
        
        questionList = li.find(class_="mt-auto d-flex jc-space-between fs-caption fc-black-400").find(class_="flex--item s-anchors s-anchors__inherit").find_all("a")
        questionToday = ""
        questionPrevious = ""
        if(len(questionList)>=2):
            questionToday = questionList[0].getText().strip()
            questionPrevious = questionList[1].getText().strip()
        
        valueDict["tagName"].append(tagName)
        valueDict["info"].append(info)
        valueDict["totalQuestion"].append(totalQuestion)
        valueDict["questionToday"].append(questionToday)
        valueDict["questionPrevious"].append(questionPrevious)
    return pd.DataFrame(valueDict)


# 个人项目/stackoverflowTags/stackoverflow tags/1859.html
dataFrameOutput = pd.DataFrame({"tagName":[], "info":[], "totalQuestion":[], "questionToday":[], "questionPrevious":[]})
for i in range(1, 1859+1):
    print(i,end=" ")
    string = "./stackoverflow tags/"+str(i)+".html"
    tmpDf = phase(string)
    dataFrameOutput = dataFrameOutput.append(tmpDf, ignore_index=True)
dataFrameOutput.to_csv("tags.csv", index=False)
```

```python
from matplotlib import pyplot as plt
tmpList = list(dataFrameOutput["totalQuestion"].values)
tagQuestions = []
for i in tmpList:
    if(" questions" in i):
        i = i.replace(" questions","")
    else:
        i = i.replace(" question","")
    i = int(i)
    tagQuestions.append(i)

tagNames = list(dataFrameOutput["tagName"].values)
tagIndex = list(range(len(tagNames)))
maxRange = 100
plt.plot(tagIndex[:maxRange], tagQuestions[:maxRange])
plt.show()
```

```python
import math
tmp = []
for i in tagQuestions:
    if(i>0):
        tmp.append(math.log(i))
    else:
        tmp.append(0)
plt.plot(tagIndex[:100], tmp[:100])
plt.show()
```
